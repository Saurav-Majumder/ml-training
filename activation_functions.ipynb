# Activation Functions

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.linspace(-10,10,10)
def plot_activation_function(f):
    vfunc = np.vectorize(f)
    y = vfunc(x)
    plt.plot(y, color='mediumseagreen')
    plt.ylabel('y')
    plt.xlabel('x')
    plt.show()
```

## Step Function

```python
step_func = lambda t: 0 if t>0 else 1
plot_activation_function(step_func)
```

## Linear Function

```python
linear_function=lambda t: 5*t
plot_activation_function(linear_function)
```

## Sigmod Function

```python
sigmod_function = lambda t: 1/(1 + math.exp(-t))
plot_activation_function(sigmod_function)
```

## Tahn Function

```python
tahn_function= lambda t: (2/(1+ math.exp(-2*t)))-1
plot_activation_function(tahn_function)
```

## Relu Function

```python
relu_function = lambda t: t if (t>0) else 0
plot_activation_function(relu_function)
```
